```python
import math
import random
import matplotlib.pyplot as plt
```

```python
def brownian_motion(step, T):
    """
    Generate a brownian motion on [0, T]
    """
    brownian = [0]
    xs = [0]
    n = int(T // step)
    for i in range(n):
        brownian.append(brownian[i] + math.sqrt(step) * random.gauss())
        xs.append(xs[i] + step)
    return xs, brownian
```

```python
for i in range(10):
    xs, brownian = brownian_motion(0.0001, 1)
    plt.plot(xs, brownian)
    plt.xlabel('time')
    plt.title('Brownian motion simulation')
    plt.legend()
plt.show()
```

```python
def brownian_bridge(step, T=1, a=0, b=0):
    """
    Generate a brownian bridge on [0, T] with X_0 = a and X_T = b
    """
    res = [0]
    xs, bridge_start = brownian_motion(step, T)
    bridge_end = bridge_start[-1]
    for i in range(1,len(xs)):
        res.append(a * (T - xs[i]) / T + b * xs[i] / T + bridge_start[i] - xs[i] / T * bridge_end)
    return xs, res 
```

```python
for i in range(10):
    xs, bridge = brownian_bridge(0.001, T=20, a=0, b=5)
    plt.plot(xs, bridge)
    plt.xlabel('time')
    plt.title('Brownian bridge simulation')
    plt.legend()
plt.show()
```
